```python
import cv2 as cv
from IPython.display import display
from PIL import Image
```

```python
img = cv.imread("cockerel.jpg")
display(Image.fromarray(img))
```

```python
img_canny = cv.Canny(img, 50, 100)
display(Image.fromarray(img_canny))
```
