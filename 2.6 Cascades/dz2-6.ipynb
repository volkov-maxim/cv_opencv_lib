Обнаружить часы с помощью каскада haarcascade_wallclock.xml

```python
import cv2 as cv
from PIL import Image
from IPython.display import display
```

```python
wallclock_classifier = cv.CascadeClassifier("haarcascade_wallclock.xml")
# wallclock_classifier = cv.CascadeClassifier("classifierWallClock.xml")
```

```python
img = cv.imread("clock1.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray = cv.equalizeHist(gray)
display(Image.fromarray(gray))
```

```python
clocks = wallclock_classifier.detectMultiScale(
    gray,
    scaleFactor=1.24,
    minNeighbors=3,
    minSize=(30, 30),
    flags = cv.CASCADE_SCALE_IMAGE
)

for x, y, w, h in clocks:
  cv.rectangle(img, (x,y), (x+w, y+h), (222, 0, 0), 2)
display(Image.fromarray(img))
```
