```python
import cv2 as cv
from PIL import Image
from IPython.display import display
```

```python
number_cascade = cv.CascadeClassifier("haarcascade_russian_plate_number.xml")
```

```python
img1 = cv.imread("number_2.jpg")
img1_gray = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
display(Image.fromarray(img1_gray))
```

```python
numbers = number_cascade.detectMultiScale(img1_gray, scaleFactor=1.1, minNeighbors=30)
for x, y, w, h in numbers:
  cv.rectangle(img1, (x, y), (x+w, y+h), (255, 0, 0), 2)
```
