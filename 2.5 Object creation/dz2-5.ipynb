Повторить изображение или придумать своё, использую все дотупные приметивы. 

```python
import numpy as np
import cv2 as cv
from PIL import Image
from IPython.display import display
```

```python
img = np.zeros((300, 300, 3), dtype=np.uint8)
img[:] = 127, 200, 10
display(Image.fromarray(img))
```

```python
img = cv.circle(img, (150, 150), 100, (256, 200, 0), cv.FILLED)
img = cv.line(img, (80, 100), (220, 100), (0, 0, 0), 3)
img = cv.rectangle(img, (100, 100), (130, 130), (0, 0, 0), cv.FILLED)
img = cv.rectangle(img, (170, 100), (200, 130), (0, 0, 0), cv.FILLED)
img = cv.putText(img, "vvvvvv", (102, 210), cv.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 2)
display(Image.fromarray(img))
```
