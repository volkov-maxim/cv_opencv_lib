```python
import numpy as np
import cv2 as cv
from PIL import Image
from IPython.display import display
```

```python
img = np.zeros((300, 300, 3), dtype=np.uint8)
display(Image.fromarray(img))
```

```python
img[:] = 119, 201, 105
display(Image.fromarray(img))
```

```python
img[100:150, 200:280] = 0, 0, 0
display(Image.fromarray(img))
```

```python
cv.rectangle(img, (50, 70), (100, 100), (0, 0, 0), 5)
display(Image.fromarray(img))
```

```python
cv.rectangle(img, (50, 130), (100, 160), (0, 0, 0), cv.FILLED)
display(Image.fromarray(img))
```

```python
cv.line(img, (0, 10), (100, 10), (0, 0, 0), 3)
display(Image.fromarray(img))
```

```python
cv.circle(img, (200, 200), 50, (0, 0, 0), cv.FILLED)
display(Image.fromarray(img))
```

```python
cv.putText(img, "  ", (100, 50), cv.FONT_HERSHEY_TRIPLEX, 1, (0, 0, 127), 1)
display(Image.fromarray(img))
```
